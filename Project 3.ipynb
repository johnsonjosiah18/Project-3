```python
from selenium import webdriver
from bs4 import BeautifulSoup

# Setup the webdriver
driver = webdriver.Chrome()

# Visit the website
url = "https://www.tax-rates.org/taxtables/property-tax-by-state"
driver.get(url)

# Create a Beautiful Soup Object
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')

# Find the div with id 'taxBox'
div = soup.find('div', id='taxBox')

# Extract all text within this div
text = div.get_text()

print(text)
```

```python
import re

# Extract all text within this div
text = div.get_text()

# Split the text by newline characters
lines = text.split('\n')
```

```python
data = []

# Iterate over each line
for line in lines:
    # Use regex to find lines that contain the state and percentage
    match = re.search(r'(\d+\.\))(.+?)(\d+\.\d+%)', line)
    if match:
        # Extract the state and percentage
        state = match.group(2).strip()
        percentage = match.group(3).strip()

        # Store in the list
        data.append([state, percentage])

# Convert the list into a DataFrame
property_tax_df = pd.DataFrame(data, columns=['State', 'Percentage'])

property_tax_df
```

```python
property_tax_df.to_csv('Property_taxes.csv', index=False)
```
