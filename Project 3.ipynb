```python
from selenium import webdriver
from bs4 import BeautifulSoup

# Setup the webdriver
driver = webdriver.Chrome()

# Visit the website
url = "https://www.tax-rates.org/taxtables/property-tax-by-state"
driver.get(url)

# Create a Beautiful Soup Object
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')

# Find the div with id 'taxBox'
div = soup.find('div', id='taxBox')

# Extract all text within this div
text = div.get_text()

print(text)
```

```python
# Load crime data csv
crime_df = pd.read_csv('estimated_crimes_1979_2022.csv')
crime_df.head()
```

```python
# Drop empty columns
crime_df.drop(columns = ['rape_revised', 'caveats'], inplace = True)
crime_df.head()
```

```python
# Rename rape column
crime_df = crime_df.rename(columns = {'rape_legacy': 'rape'})
crime_df.head()
```

```python
# Update the first row for US totals
index_to_update = 0  
crime_df.at[index_to_update, 'state_abbr'] = 'US'  # Update 'Age' in the second row
crime_df.at[index_to_update, 'state_name'] = 'United States'  # Update 'City' in the second row

crime_df.head()
```

```python
# Save the dataframe to a new csv file
crime_df.to_csv('crime_updated.csv', index=False)
```
